```python
import pandas as pd
from csv import reader

#Gaussian Naive Bayes
from sklearn import datasets
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
```

```python
#load the csv file
def load_csv(filename):
	dataset = list()
	with open(filename, 'r') as file:
		csv_reader = reader(file)
		for row in csv_reader:
			if not row:
				continue
			dataset.append(row)
	return dataset
filename = 'iris.csv'
df_list = load_csv(filename)
```

```python
#convert the list into a dataframe
df = pd.DataFrame(df_list)
df.head()
```

```python
#Removing the outcome column from the dataset but having it inluded in the target dataset
x = df.drop([4], axis=1)
y=df[4]
```

```python
#splitting the data into training and testing datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
# fit a Naive Bayes model to the data
model = GaussianNB()
model = model.fit(x_train, y_train)
```

```python
# predicted values
y_pred = model.predict(x_test)
y_pred
```

```python
# confusion matrix
Class = ['Iris-versicolor', 'Iris-setosa', 'Iris-virginica']
print(metrics.confusion_matrix(y_test, y_pred, labels = Class))
```

```python
print(metrics.classification_report(y_test, y_pred))
```
